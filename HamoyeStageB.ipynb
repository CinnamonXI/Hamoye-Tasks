```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```
 url='https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
 df= pd.read_csv(url)
 df.head()
```

```
column_names = {
    'T1': 'kitchen_temp_c', 'RH_1': '%kitchen_humidity', 'T2': 'living_room_temp_c', 'RH_2':'%living_room_humidity', 
    'T3': 'laundry_room_temp_c', 'RH_3': '%laundry_room_humidity', 'T4': 'office_room_temp_c',  'RH_4': '%office_room_humidity', 
    'T5': 'bathroom_temp_c', 'RH_5':'%bathroom_humidity', 'T6':'outside_north_side_temp_c', 'RH_6': '%outside_north_side_humidity', 
    'T7': 'ironing_room_temp_c', 'RH_7': '%ironing_room_humidity', 'T8': 'teenager_room_temp_c', 'RH_8': '%teenager_room_humidity', 
    'T9': 'parents_room_temp_c', 'RH_9': '%parents_room_humidity'
}
df= df.rename(columns=column_names)
```

```
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
scaler = MinMaxScaler()
normalised_df=df.drop(['date','lights'], axis=1)
normalised_df= pd.DataFrame(scaler.fit_transform(normalised_df), columns=normalised_df.columns)
features_df=normalised_df[['living_room_temp_c']]
target_df=normalised_df[['outside_north_side_temp_c']]
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test= train_test_split(features_df, target_df, test_size=0.3, random_state=42)
```

```
lin_model = LinearRegression()
lin_model.fit(x_train, y_train)
```

```
predict_vals = lin_model.predict(x_test)
```

```
lin_model.score(x_train, y_train)
```

Question 13

```
mae=mean_absolute_error(y_test, predict_vals)
mae
```

Question 14

```
rss= np.sum(np.square(y_test - predict_vals))
rss
```

Question 15

```
rmse=np.sqrt(mean_squared_error(y_test,predict_vals))
rmse
```
